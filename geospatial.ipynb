```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import geopandas as gpd

from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

plt.style.use('ggplot')
```

```python
africa_map = gpd.GeoDataFrame.from_file('input/Africa_SHP/Africa.shp')
africa_map.head()
```

```python
africa_map.plot(column='CODE')
```

```python
africa_pop = pd.read_csv('input/Africa_populations.tsv', sep='\t')

africa_pop_map = pd.merge(africa_map, africa_pop, left_on='COUNTRY', right_on='Country_Name')
africa_pop_map
```

```python
africa_pop_map.plot(column='Population', cmap='hot')
```
