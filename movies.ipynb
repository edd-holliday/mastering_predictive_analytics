```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

plt.style.use('ggplot')
```

```python
imbd_ratings = pd.read_csv('input/movies.csv')
```

```python
imbd_ratings.head()
```

```python
imbd_ratings.tail()
```

```python
imbd_ratings.describe()
```

```python
imbd_ratings['year'].value_counts().sort_index(ascending=False)
```

```python
imbd_ratings[(imbd_ratings['year'] > 1999) & (imbd_ratings['mpaa'] == 'R')].head()
```

```python
imbd_ratings.groupby('year')['rating'].aggregate([np.mean, np.std])
```

```python
imbd_ratings['genre'] = imbd_ratings[['Action', 'Animation', 'Comedy', 'Drama', 'Documentary', 'Romance']].idxmax(1)
```

```python
imbd_ratings['genre'].head()
```

```python
genres_map = {"Action": 'red', "Animation": 'blue', "Comedy": 'yellow', "Drama": 'green', "Documentary": 'orange', "Romance": 'purple'}

imbd_ratings['genre_colour'] = imbd_ratings['genre'].apply(lambda x: genres_map[x])

imbd_ratings['genre_colour'].head()
```

```python
pd.pivot_table(imbd_ratings, values='rating', index='year', columns=['genre'], aggfunc=np.mean)
```

```python
imbd_ratings['length'].plot()
```

```python
plot1 = imbd_ratings.groupby('genre')['length'].plot(kind='kde', xlim=(0,120), legend='year')
plot1[0].set_xlabel('Number of Minutes')
plot1[0].set_title('Distribution of Film Genre by Runtime Minutes')
```

```python
pd.pivot_table(imbd_ratings, values='rating', index='title', columns=['genre']).plot(kind='box').set_ylabel('Rating')
```

```python
pd.plotting.scatter_matrix(imbd_ratings[['year', 'budget', 'rating', 'votes']], alpha=0.2, figsize=(6,6), diagonal='kde')
```

```python
fig, ax = plt.subplots(nrows=3, ncols=2, figsize=(15,15))

row, col = 0, 0

for index, genre in imbd_ratings.groupby('genre'):
    if row > 2:
        row = 0
        col += 1

    genre.groupby('genre').plot(ax=ax[row, col], kind='scatter', x='length', y='rating',
                                 s=np.sqrt(genre['votes']), c=genre['genre_colour'],
                                 xlim=(0,120), ylim=(0,10),
                                 alpha=0.5, label=index)
    row += 1
```
